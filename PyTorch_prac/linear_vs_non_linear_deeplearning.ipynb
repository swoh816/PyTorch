```python
import torch as tc
import numpy as np
import random as rd
import matplotlib.pyplot as plt

from torch.autograd import Variable
```

# Nonlinear data prediction

```python
x_data = [10*rd.random()-5 for i in range(1000)]
y_data = [i**2 for i in x_data]
```

```python
plt.scatter(x_data, y_data)
```

```python
plt.show()
```

```python
x_data
```

```python
x_data_var = [[i] for i in x_data]
```

```python
x_data_var = Variable(tc.Tensor(x_data_var), requires_grad=True)
```

```python
y_data_var = [[i] for i in y_data]
```

```python
y_data_var = Variable(tc.Tensor(y_data_var))
```

```python
class nonlinear(tc.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = tc.nn.Linear(1,1)
        self.relu = tc.nn.functional.relu
        
    def forward(self, x):
        return(self.relu(self.linear(self.relu(self.linear(self.relu(self.linear(x)))))))

    
class linear(tc.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = tc.nn.Linear(1,1)
        
    def forward(self, x):
        return(self.linear(x))
```

```python
nonlin_model = nonlinear()
criterion = tc.nn.MSELoss()
optimizer = tc.optim.SGD(nonlin_model.parameters(), lr=0.01)
```

```python
for epoch in range(1000):
    y_pred = nonlin_model(x_data_var)
    loss = criterion(y_pred, y_data_var)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
x_test = tc.Tensor([[10*rd.random()-5] for i in range(1000)])

y_test_nonlin = nonlin_model(Variable(x_test))
```

```python
plt.scatter(x_test,y_test_nonlin)
```

```python
plt.show()
```

```python
lin_model = linear()
criterion = tc.nn.MSELoss()
optimizer = tc.optim.SGD(lin_model.parameters(), lr=0.01)
```

```python
for epoch in range(1000):
    y_pred = lin_model(x_data_var)
    loss = criterion(y_pred, y_data_var)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
y_test_lin = lin_model(Variable(x_test))
```

```python
plt.scatter(x_test,y_test_lin)
```

```python
plt.show()
```

```python
plt.scatter(x_test, y_test_nonlin, label='Nonlin')
plt.scatter(x_test, y_test_lin, label='Lin')
plt.scatter(x_data, y_data, label='Original')
plt.legend()
plt.show()
```
